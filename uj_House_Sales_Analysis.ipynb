```python
import os
import pandas as pd

# Print current working directory
print("Current working directory:", os.getcwd())

# List contents of the Downloads folder
downloads_path = "/Users/nemethlaszlo/Downloads"
print("Contents of Downloads:", os.listdir(downloads_path))

# CSV file path
csv_path = "/Users/nemethlaszlo/Downloads/kc_house_data.csv"

# Check if the file exists and load the dataset if it does
if os.path.exists(csv_path):
    print("File exists! Loading the dataset...")
    data = pd.read_csv(csv_path)
    # Print dataset information
    print(data.info())
    print(data.head())
else:
    print("File not found at:", csv_path)
```

#### Exploratory Data Analysis (EDA)

##### Descriptive Statistics
This section provides an overview of the dataset’s key statistics, helping us understand the distribution and central tendencies of the variables.

```python
# Calculate descriptive statistics
desc_stats = data.describe()
print(desc_stats)
```

##### Data Visualizations
The following charts illustrate the relationships between property features and sale prices. These insights help us identify trends and potential pricing strategies.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Distribution of Sale Prices
plt.figure(figsize=(10,6))
sns.histplot(data['price'], bins=50, kde=True)
plt.title("Distribution of House Sale Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# Example: Relationship between Square Footage and Price
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['sqft_living'], y=data['price'])
plt.title("House Sale Price vs. Square Footage")
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.show()

# Additional visualizations can be added here.
```

#### Analysis and Business Insights

Based on the exploratory analysis, several key insights emerge:

- **Square Footage:** There is a strong positive correlation between the square footage of a property and its sale price. Larger houses tend to command higher prices.
- **Bedrooms/Bathrooms:** The number of bedrooms and bathrooms also significantly affects the sale price, indicating that properties with more amenities may be priced higher.
- **Other Factors:** Additional variables such as property condition and age might also play a role in determining its value.

**Business Recommendations:**  
- **Pricing Strategy:** Consider implementing tiered pricing models based on square footage and property amenities.
- **Targeted Marketing:** Focus on properties with unique features or those that significantly deviate from the average trend.

```python
# (Optional advanced analysis, e.g., linear regression)
import statsmodels.api as sm

# Define features and target variable
X = data[['sqft_living', 'bedrooms', 'bathrooms']]
y = data['price']

# Add constant term for intercept
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()
print(model.summary())
```
